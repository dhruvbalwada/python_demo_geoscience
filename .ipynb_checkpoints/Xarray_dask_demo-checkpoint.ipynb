### Xarray 

http://xarray.pydata.org/

Xarray is a python library that can be used to handled labeled array data. 
It is like pandas, but for larger dimensional arrays. It follows the model of netcdf, and brings it to Python and allows quite easily to extend it to other sets. 

```python
import xarray as xr 
from matplotlib import pyplot as plt
import seaborn as sns
from xmitgcm import open_mdsdataset

%matplotlib inline
```

<img src=http://xarray.pydata.org/en/stable/_images/dataset-diagram.png>

Each data set has values, dimensions, coordinates and attributes.

```python
# A simple sine function example 

xdim = np.array(range(0,100))
dx = 2*np.pi/100
xcoord = xdim*dx

sine = np.sin(xcoord)

plt.plot(sine)
```

```python
sine_xr = xr.DataArray(sine, coords=[xcoord], dims='x')
sine_xr
```

```python
print sine_xr.dims 
print sine_xr.coords
```

```python
sine_xr.plot()
```

```python
print sine_xr.max(), '\n', sine_xr.min()
```

```python
sine_xr.isel(x=10)
```

```python
sine_xr.sel(x=4, method='nearest')
```

```python
tdim = np.array(range(0,100))
dt = 0.1
tcoord = tdim*dt

cosine = np.cos(tcoord)
```

```python
cosine_xr = xr.DataArray(cosine, coords=[tcoord], dims='t')
cosine_xr.plot()
```

```python
sincos_xr = sine_xr*cosine_xr
```

```python
sincos_xr
```

```python
sincos_xr.plot()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xg,tg = np.meshgrid(sincos_xr.x, sincos_xr.t)
ax.plot_surface(tg, xg, sincos_xr.values,cmap='RdBu_r')
plt.xlabel('t')
plt.ylabel('x')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xg,tg = np.meshgrid(sincos_xr.x, sincos_xr.t)
ax.contour(tg, xg, sincos_xr.values,cmap='RdBu_r')
plt.xlabel('t')
plt.ylabel('x')
```

Allows us to do Matlaby things, with simple syntax. Also (IMHO) Python is much better at visualization compared to Matlab. 

### Let's look at a slightly bigger dataset. 

Xarray does lazy evaluation, which implies that the data will not be loaded into the memory unless a computation needs to done. 

```python
ds = open_mdsdataset(ddir, delta_t=300, prefix=['U','V','T','W','S','Eta']
                     ,ignore_unknown_vars=True, geometry='cartesian')
```

```python
ds['U'].attrs
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(111)
%time ds['T'].isel(Z=0,time=-20).plot(vmin=2.5, vmax=6.5)
ax.set_aspect('equal')
plt.tight_layout()
```

```python
%time ds['T'].isel(time=slice(0,10)).mean(['time','XC','YC']).plot()
%time ds['T'].isel(time=slice(0,400)).mean(['time','XC','YC']).plot()
```

```python
ds['T'].isel(time=slice(0,400)).mean(['time','XC']).plot.contourf()
```
